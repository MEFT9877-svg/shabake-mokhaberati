```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Import seaborn for better styling

# Sample data - Replace with your actual data
training_steps = np.linspace(0, 200000, 100)

# Dummy data - REPLACE with your actual data.  Use numpy.normal for more realistic distributions
standard_episode_length = np.random.normal(0.8, 0.1, 100).clip(0, 1)
comfort_episode_length = np.random.normal(0.4, 0.1, 100).clip(0, 1)
aggressive_episode_length = np.random.normal(0.9, 0.05, 100).clip(0, 1)

standard_high_speed_reward = np.random.normal(-10, 5, 100)
comfort_high_speed_reward = np.random.normal(-5, 3, 100)
aggressive_high_speed_reward = np.random.normal(0, 7, 100)

standard_right_lane_reward = np.random.normal(0.8, 0.1, 100).clip(0, 1)
comfort_right_lane_reward = np.random.normal(0.5, 0.15, 100).clip(0, 1)
aggressive_right_lane_reward = np.random.normal(0.2, 0.05, 100).clip(0, 1)

standard_collision_reward = np.random.normal(2, 3, 100).clip(0, 20)
comfort_collision_reward = np.random.normal(7, 4, 100).clip(0, 20)
aggressive_collision_reward = np.random.normal(15, 5, 100).clip(0, 20)

# Plotting with Seaborn styling and improved layout
sns.set_style("darkgrid")  # Set a nice grid style
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

driving_styles = ["Standard", "Comfort", "Aggressive"]

plot_data = [
    (axes[0, 0], "Episode Length (Normalized 0 to 1)", [standard_episode_length, comfort_episode_length, aggressive_episode_length]),
    (axes[0, 1], "High Speed Reward", [standard_high_speed_reward, comfort_high_speed_reward, aggressive_high_speed_reward]),
    (axes[1, 0], "Right-most Lane Reward", [standard_right_lane_reward, comfort_right_lane_reward, aggressive_right_lane_reward]),
    (axes[1, 1], "Collision Reward", [standard_collision_reward, comfort_collision_reward, aggressive_collision_reward]),
]

for ax, title, data in plot_data:
    for i, style_data in enumerate(data):
        ax.plot(training_steps, style_data, label=driving_styles[i])
    ax.set_title(title, fontsize=14)  # Increased title font size
    ax.set_xlabel("Training Steps", fontsize=12)  # Increased x-label font size
    ax.set_ylabel(title.split("(")[0].strip(), fontsize=12) # Use title without normalization as ylabel
    ax.legend(fontsize=10)  # Increased legend font size
    ax.grid(True)
    ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0)) # Scientific notation for x-axis if needed


plt.tight_layout(padding=1.5)  # Add padding around subplots for better spacing
plt.show()
```
